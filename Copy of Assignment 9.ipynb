Conditional Statement In Python


Example 01. Only If Or If Condition

```python
age = 20

if age >= 20:
    print("You are a adult")

```

Example 2. If....Else


```python
age = 80
if age >= 80:
    print("You are a adult")
else:
    print("You are not a adult")
```

```python
marks = 90
if marks >= 120:
    print("Grade: A")
elif marks >= 85:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Fail")

```
